# Line graph

```
# To plot line graph, we need time series data. Now, letâ€™s first create time series data.
# So here, we create  five hundred random data and index with date for period of 500 days. 
# The method cumsum stands for cumulative sum. Thus for every new number randomly generated, the function adds it to the previous number. 

import pandas as pd
import numpy as np
Ts = pd.Series(np.random.randn(500), index=pd.date_range("1/1/2000", periods=500))
Ts = Ts.cumsum()
```

```
# And finally, we use plot method to generate such time series graph.
# By running this line of code results in follwing time series graph.

Ts.plot()
```

# Scatter plot

```
# To plot scatter plot, we need a dataframe with two columns. 
# Assume, we have a dataframe with two columns, in this case column A and B.
# These columns contains fifty random data. So, we have a dataframe with two columns and fifty rows. 

import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.rand(50, 1), columns=['A'])
df['B']= pd.Series(list(range(50)))
```

```
# Now, for plotting to scatter plot using pandas there is DataFrame class and this class has a member called plot.
# Calling the scatter() method on the plot member draws a plot between two columns of pandas DataFrame. 
# Here, we also define axis label, label A to x axis and label B to y axis.
# By running this line of code results in following scatter plot.

df.plot.scatter(x='A', y='B')
```

# Histogram


```
# To create histogram, let us first create some data. 
# Assume, we created 100 random numbers for column 'a', and 'b'. 

import pandas as pd
import numpy as np
df = pd.DataFrame({'a': np.random.randn(100), 'b': np.random.randn(100)})
```

```
# After that, we use hist method of plot member of dataframe class to see how frequently each datapoint occurs. 
# By running this line of code, you can see, the x-axis shows the different numbers, and the y-axis shows how frequently they occur in the data.
# Here in the plot, the orange bar indicates the column 'b,' and the blue bar indicates column 'a.'

df.plot.hist()
```

# Correlation Plot

```
# To demonstrate the example of correlation plot, assume we have a following dataframe which contains three columns column A, B and C. 
# Here, each column contains random datapoint.

import pandas as pd
import numpy as np
df = pd.DataFrame(list(range(50)), columns=['a'])
df['b'] = pd.Series(np.random.rand(50))
df['c'] = pd.Series(np.random.rand(50))
```

```
# Now we add some correlation to column c and b.

df['c'] = df['c'] + 0.5 * df['a']
df['b'] = df['b'] - 0.5 * df['a']
```

```
# To check the Correlation matrix, we use the Correlation method.
# Running this line of code results in the Correlation matrix.
# As you can see, the diagonal elements are positively correlated and off- diagonal are negatively correlated.

df.corr()
```

```
# To plot correlation polt, we use another library called matplotlib. 
# We import this library as import matplotlib dot pyplot as p l t.
# After that we use matshow function and pass correlation dataframe.
# Running this line of code results in following correltion matrix. 
# Here, in the correlation matrix, the yellow color denotes positive correlation and blue color denotes negative correlation.

import matplotlib.pyplot as plt
plt.matshow(df.corr())
```
